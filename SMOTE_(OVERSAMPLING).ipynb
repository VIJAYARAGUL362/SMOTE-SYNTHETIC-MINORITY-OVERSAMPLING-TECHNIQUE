IMPORTING THE LIBRARIES


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

CREATING THE LIBRARIES

```python
dataset = {
    "ANIMAL":[0,0,0,1,0,1],
    "HEIGHT":[1,2,2.5,0.6,1.5,0.9],
    "WEIGHT":[5,10,15,2.5,7,1.5]
}


DATAFRAME = pd.DataFrame(dataset)
DATAFRAME

X = DATAFRAME.iloc[:,1:]
y = DATAFRAME.iloc[:,0]
```

***VISUALIZING THE DATASET***

this is to find whether the dataset is balanced or unbalanced

```python
sns.countplot(data=DATAFRAME,x="ANIMAL",hue="ANIMAL",stat="count")
```

VISUALIZING THE DATASET

```python
sns.scatterplot(data = DATAFRAME, x = "WEIGHT",y = "HEIGHT",hue="ANIMAL" )
```

APPLYING SMOTE ON THE DATASET

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42,k_neighbors=1,sampling_strategy="minority")
x_new,y_new = sm.fit_resample(X,y)


NEW_DATAFRAME = pd.concat([x_new,y_new],axis=1)
NEW_DATAFRAME = NEW_DATAFRAME.round(2)
NEW_DATAFRAME
```

VISUALIZING AFTER THE SMOTE's METHOD IS APPLIED

```python
sns.scatterplot(data = NEW_DATAFRAME, x = "WEIGHT",y = "HEIGHT",hue="ANIMAL" )
```
